# Inference with Torch

## 导包

```python
import sys
sys.path.append('../')
from unet import UNet

import numpy as np
import os 
from matplotlib import pyplot as plt
from skimage import io,transform
from PIL import Image

import torch
```

## 读取图片

```python
img = Image.open(os.path.join("images", "input.jpg"))
print(type(img))
print(img.size)
plt.imshow(img)
```

#### PIL -> ndarray

```python
img = np.array(img)
print(type(img))
print(img.shape)
print(img[0][0])
print(img.dtype)
```

#### 换dims & 增加dims

```python
img = img.transpose((2, 0, 1))
print(type(img))
print(img.shape)
print(img[0][0])
print(img.dtype)
```

```python
img = np.expand_dims(img, axis=0)
print(type(img))
print(img.shape)
print(img[0][0])
print(img.dtype)
```

#### uint8 -> float

```python
if img.max() > 1:
    img = img / 255

print(type(img))
print(img.shape)
print(img[0][0])
print(img.dtype)
```

```python
img = img.astype(np.float32)
print(type(img))
print(img.shape)
print(img[0][0])
print(img.dtype)
```

#### ndarray -> tensor

```python
img_tensor = torch.from_numpy(img)
print(type(img_tensor))
print(img_tensor.shape)
print(img_tensor[0][0])
print(img_tensor.dtype)
print(img_tensor.device)
```

```python
img_tensor = img_tensor.to(device="cuda:0")
print(img_tensor.device)
```

## 准备模型

```python
net = UNet(n_channels=3, n_classes=1)
net.load_state_dict(torch.load("../checkpoints/CP_epoch100.pth"))
net.cuda()
net
```

## 推理

```python
%%time
result = net(img_tensor)
```

```python
print(type(result))
print(result.shape)
print(result[0][0])
print(result.dtype)
print(result.device)
```

## 显示图片

```python
result_numpy = result.cpu().detach().numpy()
print(type(result_numpy))
print(result_numpy.shape)
print(result_numpy.dtype)
```

```python
res = np.array(result_numpy[0][0])
print(type(res))
print(res.shape)
```
